```python
from sklearn.datasets import load_boston, make_regression
from sklearn.cross_validation import train_test_split
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X, y = make_regression(10000, n_features=20)
aX, bX, aY, bY = train_test_split(X,y)
```

```python
gb1 = GradientBoostingRegressor(n_estimators=100, sub_sample_size=0.5)
%time gb1.fit(aX,aY)
print "gb1 :", gb1.score(bX,bY)
plt.plot(gb1.score_by_depth(bX,bY))
plt.show()

gb2 = GradientBoostingRegressor(n_estimators=100, max_depth = 4, max_features=10)
%time gb2.fit(aX,aY)
print "gb2 :", gb2.score(bX,bY)
plt.plot(gb2.score_by_depth(bX,bY))
plt.show()

gb3 = SKG(n_estimators=100, max_depth = 3)
%time gb3.fit(aX,aY)
print "gb3 :", gb3.score(bX,bY)
print "gb3 :", RMSE(gb3.predict(bX),bY)
```
